```python
import pandas as pd
Data = pd.read_excel(r'class_grades.xlsx')
class Section:

    def __init__(self):
        # declaration of variables
        self.LNames = list()
        self.Names = list()
        self.GPrelims = list()
        self.GMidterms = list()
        self.GFinals = list()
        self.Semestral = list()
        self.fail = list()
        self.simStudents = list()

        # insertion of the data from the xlsx file to the table
        self.appendN = {self.LNames.append(i) for i in Data['name']}
        self.appendP = {self.GPrelims.append(i) for i in Data['prelim_grades']}
        self.appendM = {self.GMidterms.append(i) for i in Data['midterm_grades']}
        self.appendF = {self.GFinals.append(i) for i in Data['final_grades']}
        self.appendS = {self.Semestral.append( self.GPrelims[i] * 0.3 + self.GMidterms[i] * 0.3 +
                        self.GFinals[i] * 0.4) for i in range(len(self.LNames))}
        
        # convertion of list to dictionary
        self.Pdict = {self.LNames[i]: self.GPrelims[i] for i in range(len(self.LNames))}
        self.Mdict = {self.LNames[i]: self.GMidterms[i] for i in range(len(self.LNames))}
        self.Fdict = {self.LNames[i]: self.GFinals[i] for i in range(len(self.LNames))}
        self.Sdict = {self.LNames[i]: self.Semestral[i] for i in range(len(self.LNames))}

        # list/set comprehension
        self.grades = [[self.LNames[i], self.Pdict[self.LNames[i]], self.Mdict[self.LNames[i]],
                        self.Fdict[self.LNames[i]], self.GPrelims[i] * 0.3 + self.GMidterms[i] * 0.3 +
                        self.GFinals[i] * 0.4] for i in range(len(self.LNames))]
        self.sorted_Plist = sorted([[self.grades[i][0], self.grades[i][1]]
                    for i in range(len(self.grades))], key=lambda x: x[1])

        self.sorted_Mlist = sorted([[self.grades[i][0], self.grades[i][2]] 
                    for i in range(len(self.grades))],key=lambda x: x[1])
            
        self.sorted_Flist = sorted([[self.grades[i][0], self.grades[i][3]] 
                    for i in range(len(self.grades))],key=lambda x: x[1])   
            
        self.sorted_Slist = sorted([[self.grades[i][0], self.grades[i][4]] 
                    for i in range(len(self.grades))],key=lambda x: x[1]) 
        
    def get_grades(self):
        print("{:<10} {:<15} {:<10} {:<10} {:<10} {:<10}".
        format('Rank', 'Name', 'Prelim', 'Midterm', 'Finals', 'Semestral Grade'))
        self.increment = 0

        for grade in self.grades:
            Name, Prelim, Midterm, Finals, SemestralGrade = grade
            print('  {}'.format(self.increment + 1), "{:<8} {:<15} {:<10} {:<10} {:<10} {:<10}"
                    .format('', Name, Prelim, Midterm, Finals, SemestralGrade))
            self.increment += 1

    def get_class_stats(self):

        Class_stats={}               
        Class_stats['Prelim Lowest      ']= self.sorted_Plist[0][1]   
        Class_stats['Prelim Highest     ']= self.sorted_Plist[len(self.sorted_Plist)-1][1]  
        Class_stats['Midterms Lowest    ']= self.sorted_Mlist[0][1]
        Class_stats['Midterms Highest   ']= self.sorted_Mlist[len(self.sorted_Plist)-1][1]  
        Class_stats['Finals Lowest      ']= self.sorted_Flist[0][1]
        Class_stats['Finals Highest     ']= self.sorted_Flist[len(self.sorted_Plist)-1][1]
        Class_stats['Semestral Lowest   ']= self.sorted_Slist[0][1]
        Class_stats['Semestral Highest  ']= self.sorted_Slist[len(self.sorted_Plist)-1][1]

        Class_stats['Mean of Prelim     ']=(sum(self.GPrelims)) / (len(self.GPrelims))
        Class_stats['Mean of Midterm    ']=(sum(self.GMidterms))/ (len(self.GMidterms))
        Class_stats['Mean of Finals     ']=(sum(self.GFinals))  / (len(self.GFinals))
        Class_stats['Mean of Semestral  ']= (sum(self.Semestral))  / (len(self.Semestral))

        Class_stats['Median of Prelim   ']= self.sorted_Plist[(round(
                                                        (len(self.GPrelims))/2)-1)][1]
        Class_stats['Median of Midterm  ']= self.sorted_Mlist[(round(
                                                        (len(self.GMidterms))/2)-1)][1]
        Class_stats['Median of Finals   ']= self.sorted_Flist[(round(
                                                        (len(self.GFinals))/2)-1)][1]
        Class_stats['Median of Semestral']= self.sorted_Slist[(round(
                                                        (len(self.Semestral))/2)-1)][1]

        Class_stats['Mode of Prelim     ']=max(self.GPrelims, key = self.GPrelims.count)
        Class_stats['Mode of Midterm    ']=max(self.GMidterms, key = self.GMidterms.count)
        Class_stats['Mode of Finals     ']=max(self.GFinals, key = self.GFinals.count)
        Class_stats['Mode of Semestral  ']=max(self.Semestral, key = self.Semestral.count)

        Class_stats['Standard Deviation of Prelim    ']=   (sum(pow(x-(sum(self.GPrelims))/ 
                                                (len(self.GPrelims)),2) for x in self.GPrelims) 
                                                / len(self.GPrelims))**(1/2)

        Class_stats['Standard Deviation of Midterm   ']=   (sum(pow(x-(sum(self.GMidterms))/ 
                                                (len(self.GMidterms)),2) for x in self.GPrelims) 
                                                / len(self.GMidterms))**(1/2)

        Class_stats['Standard Deviation of Finals    ']=   (sum(pow(x-(sum(self.GFinals))/ 
                                                (len(self.GFinals)),2) for x in self.GPrelims) 
                                                / len(self.GFinals))**(1/2)
                                                
        Class_stats['Standard Deviation of Semestral ']=    (sum(pow(x-(sum(self.Semestral))/ 
                                                (len(self.Semestral)),2) for x in self.Semestral) 
                                                / len(self.Semestral))**(1/2)   
        print("The Class Data " , Class_stats) 

    def get_similar_students(self):
        similar = [x for n, x in enumerate(self.Semestral) if x in self.Semestral[:n]]
        reverse = {value : key for (key, value) in self.Sdict.items()}
        for i in reverse:
            for x in similar:
                if str(i) == str(x):
                    self.simStudents.append(reverse.get(x))
                    reverse[i] = reverse.pop(i)
                else:
                    pass

        print('The similar students are :',list(dict.fromkeys(self.simStudents)))

    def get_failed(self):
        reverse = {value : key for (key, value) in self.Sdict.items()}
        
        for i in range(len(self.sorted_Slist)):
            if int(self.sorted_Slist[i][1]) <= 70:
                self.fail.append(self.sorted_Slist[i][0])
            else:
                pass
        print("The students that failed are:", self.fail)

    def failed_count(self):
        print("The students count that failed class :", (len(self.fail)))

#call one by one
section = Section()
section.get_grades()
section.get_class_stats()
section.get_similar_students()
section.get_failed()
section.failed_count()
```
